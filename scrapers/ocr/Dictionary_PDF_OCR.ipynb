```python
from google.colab import files
# Download the pdf from https://www.memoriachilena.gob.cl/archivos2/pdfs/MC0049597.pdf
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]  # Get the first (and only) uploaded file name
print(f"Uploaded file: {pdf_file}")
```

```python
import os
from pdf2image import convert_from_path
import pytesseract

# Output text file to store the OCR results
output_text = 'ocr_output.txt'

print("Converting PDF pages to images...")
pages = convert_from_path(pdf_file, dpi=300)
print(f"Total pages converted: {len(pages)}")

# Write the OCR results to the output file
with open(output_text, 'w', encoding='utf-8') as f:
    for i, page in enumerate(pages, start=1):
        print(f"Processing page {i}...")
        text = pytesseract.image_to_string(page)
        f.write(f"--- Page {i} ---\n")
        f.write(text)
        f.write("\n\n")

print(f"OCR complete! Output saved to {output_text}")
```
